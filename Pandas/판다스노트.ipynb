#### indexing, slicing, 조건 필터링

```python
# loc - indexing/slicing
# indexing과 slicing 할 수 있다
# slicing은 [시작(포함):끝(포함)] 이다. 둘 다 포함!!
# df.loc[행,열]
df.loc[5,'class']  # index 5행, class 열의 값 출력
```

```python
# 조건 필터
cond = df['age'] >= 70 # 'age' 열의 값들 중 70 이상인 것들 (True)들
df.loc[cond] #True인것들 출력
```

```python
# loc 다중 조건

# 조건 1 정의
cond1 = df['fare'] > 30

# 조건 2 정의
cond2 = df['who'] == 'woman' 

df.loc[cond1 & cond2]
df.loc[cond1 | cond2]
```

```python
# 조건 필터 후 데이터 대입하기
cond = (df['age'] >= 70)
cond
df.loc[cond] # 'age'열의 값이 70이상인 행들 출력
# 나이 컬럼만 가져오기
df.loc[cond,'age'] # 'age'열의 값이 70이상인 행들 중 'age'컬럼만 가져오기 df.loc[행,열]
# 가져온 컬럼에 -1값 넣기
df.loc[cond,'age'] = -1
df.loc[cond]
```

```python
# iloc
# loc와 유사, 하지만 index만 허용
df.head()
df.iloc[1,3] # df.iloc[행index, 열index]
df.iloc[[0,3,4],[0,1,5,6]] #df.iloc[ [행 리스트] , [열 리스트] ]
```

# 06. Group by 와 Pivot table
- groupby()는 데이터를 피봇팅하여 통계량을 볼 수 있도록 도와주는 메서드이면서, 데이터를 특정 조건에 맞게 전처리해 줄 때 용이
- pivot_table()은 데이터를 특정 조건에 따라 행(row)과 열(column)을 기준으로 데이터를 펼쳐서 그에 대한 통계량을 볼 때 활용

```python
from IPython.display import Image
import numpy as np
import pandas as pd
import seaborn as sns


df = sns.load_dataset('titanic')
df.head()
```

- apply() 함수를 적용


```python
df.head()
```

```python
df['who'].value_counts()
```

```python
def transform_who(x):
    if x == 'man':
        return '남자'
    elif x == 'woman':
        return '여자'
    else:
        return '아이'

df['who'].apply(transform_who)
```

```python
df['who'].apply(transform_who).value_counts()
```

```python
def transform_who(x):
    return x['fare'] / x['age']
df.apply(transform_who, axis=1)
```

- apply() -lambda 함수

```python
df['survived'].value_counts()
```

```python
df.head()
```

```python
df['survived'].apply(lambda x: '생존' if x == 1 else '사망')
```

- Group by()

```python
df.head()
```

```python
df.groupby('sex').mean(numeric_only=True)
```

```python
df.groupby( ['sex','pclass']).mean(numeric_only=True)
```

```python
# 만약 survived 열만 출력하고 싶다면
# 표이름.groupby([행 리스트])[[열 이름 리스트]].집계함수()
df.groupby( ['sex','pclass'] )[['survived']].mean()
```

- reset_index() : 인덱스 초기화
    - 그룹핑된 데이터프레임의 index를 초기화하여 새로운 데이터프레임을 생성한다.
    

```python
# index 초기화
df.groupby(['sex','pclass'])['survived'].mean().reset_index()
```

```python
# 성별, 좌석등급 별 통계
df.groupby(['sex', 'pclass'])[['survived', 'age']].agg(['mean','sum'])
```

```python
df.groupby(['sex', 'pclass'])[['survived', 'age']].agg(['mean','sum']).reset_index()
```

### 정리
- groupby 연산자 + 집계함수()
    + 표이름.groupby( [ 행 리스트 ]) [[열 리스트]].집계함수()
    + reset_index() 인덱스 초기화

- pivot_table()
    + index, columns, values을 지정해 피벗

```python
# index에 그룹을 표기
df.pivot_table(index='who', values = 'survived')
```

```python
# columns에 그룹을 표기
df.pivot_table(columns='who', values='survived')
```

```python
df.pivot_table(index=['who','pclass'],values='survived')
```

```python
df.pivot_table(index='who', columns='pclass',values='survived')
```

# 07. 연결(concat)과 병합(merge)
- concat() : 2개 이상의 데이터프레임을 행 혹은 열 방향으로 연결
- merge() : 2개의 데이터프레임을 특정 key를 기준으로 병합할 때 활용하는 메서드

```python
df1 = pd.DataFrame({
    '고객명': ['박세리', '이대호', '손흥민', '김연아', '마이클조던'],
    '생년월일': ['1980-01-02', '1982-02-22', '1993-06-12', '1988-10-16', '1970-03-03'],
    '성별': ['여자', '남자', '남자', '여자', '남자']})
df1
```

```python
df2 = pd.DataFrame({
    '고객명': ['김연아', '박세리', '손흥민', '이대호', '타이거우즈'],
    '연봉': ['2000원', '3000원', '1500원', '2500원', '3500원']})
df2
```

```python
# 행 방향으로 연결 axis = 0 (위아래로), 중복값 알아서 제거됨
# index값 그대로 유지지
pd.concat([df1,df2])
```

```python
# 인덱스 합쳐서 매겨짐
df3 = pd.concat([df1,df2],ignore_index=True)
df3
```

```python
# 열 방향으로 연결
pd.concat([df1,df2], axis=1)
```

- merge() : 병합
    + 서로 다른 구성의 데이터프레임이지만, 공통된 key 값(컬럼)을 가지고 있다면 병합할 수 있습니다.

```python
df1
```

```python
df2
```

```python
pd.merge(df1,df2)
```

- 병합하는 방법 4가지
    + how옵션 이용
        + how : {left, right, outer, inner}
    + 디폴트값은 inner


```python
# 병합하는 방법 4가지 기본 inner : 공통된 칼럼만 표시시
pd.merge(df1,df2)
```

```python
pd.merge(df1,df2, how='left')
```

```python
pd.merge(df1,df2,how='right')
```

```python
pd.merge(df1,df2,how='outer')
```
